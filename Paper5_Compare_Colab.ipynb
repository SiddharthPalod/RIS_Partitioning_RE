## 3) Run compare (train + visualize)

You can tune `STEPS`, `SEED`, `LAMBDA1`, `LAMBDA2`.

```python
import subprocess

STEPS = 8000
SEED = 42
LAMBDA1 = 0.6
LAMBDA2 = 0.008

cmd = [
    'python', 'compare.py',
    '--seed', str(SEED),
    '--steps', str(STEPS),
    '--lambda1', str(LAMBDA1),
    '--lambda2', str(LAMBDA2)
]
print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=True)
print('Done.')
```

## 4) Show comparison figure

```python
from IPython.display import Image, display
display(Image(filename='results/compare.png'))
```

## 5) Show all generated figure panels (if present)

```python
from pathlib import Path
from IPython.display import display, Image

fig_dir = Path('results/figures')
if not fig_dir.exists():
    print('No results/figures directory found yet.')
else:
    pngs = sorted(fig_dir.glob('*.png'))
    if not pngs:
        print('No PNG figures found in results/figures.')
    else:
        for p in pngs:
            print(p.name)
            display(Image(filename=str(p)))
```
